## Exploration of Number of Visits per Patient

How many times does a typical patient visit the ED?
- ~29% of patients have visited the ED more than once in the period and ~71% visited only once

Are there "frequent flyers" in the ED? What do these patients look like?

What is the average length of stay in the ED?

Do patients return for the same issues or different issues?

```python
visit_counts = visits.MRN.value_counts().reset_index()
visit_counts.columns = ['MRN', 'visit_count']
visit_counts
```

```python
# Determine how many people visit more than once
visit_counts['more_than_once'] = visit_counts['visit_count'] > 1
return_visitors = visit_counts['more_than_once'].value_counts().reset_index()
total = return_visitors['count'].sum()
return_visitors['percent'] = (return_visitors['count'] / total * 100).round(1)
return_visitors
```

```python
# Explore frequent flyers

# Create bins to explore the data
bins = np.arange(1, 111, 5)  # 1-10, 11-20, ..., 101-110

# Label each bin
labels = [f'{i}-{i+4}' for i in bins]

# Add binned column
visit_counts['visit_bin'] = pd.cut(visit_counts['visit_count'], 
                                   bins=np.append(bins, np.inf), 
                                   right=False, labels=labels)

bin_counts = visit_counts['visit_bin'].value_counts().sort_index()

bin_total = bin_counts.sum()
bin_percentages = (bin_counts / bin_total * 100).round(2)
bin_percentages

# 97% have fewer than 5 visits
```

```python
# Define people with more than 50 visits as FF

ff_mrns = visit_counts[visit_counts['visit_count'] > 50]['MRN']
ff_visits = visits[visits['MRN'].isin(ff_mrns)]

cc_summary = (
    ff_visits.groupby(['MRN', 'CC'])
    .size()
    .reset_index(name='cc_count')
    .sort_values(['MRN', 'cc_count'], ascending=[True, False])
)

cc_summary
```

```python
dc_summary = (
    ff_visits.groupby(['MRN', 'DC_dispo'])
    .size()
    .reset_index(name='dc_dispo_count')
    .sort_values(['MRN', 'dc_dispo_count'], ascending=[True, False])
)

dc_summary
```

```python
# Look at the top patient that came in 100+ times
top_mrn = visit_counts.sort_values('visit_count', ascending=False).iloc[0]['MRN']
top_patient_visits = visits[visits['MRN'] == top_mrn].copy()

# Parse arrival time as datetime
top_patient_visits['Arrival_time'] = pd.to_datetime(top_patient_visits['Arrival_time'])

# Sort visits chronologically
top_patient_visits = top_patient_visits.sort_values('Arrival_time')

# Create 'days since first visit'
start_time = top_patient_visits['Arrival_time'].iloc[0]
top_patient_visits['days_since_first'] = (top_patient_visits['Arrival_time'] - start_time).dt.days
```

```python
# Filter out all categories less than 1%
# Step 1: Compute % distribution of DC_dispo
dc_counts = visits['DC_dispo'].value_counts(normalize=True)  # gives proportion

# Step 2: Identify categories with <1% (i.e., <0.01)
rare_categories = dc_counts[dc_counts < 0.001].index

# Step 3: Replace those rare categories with "Other"
visits['DC_dispo_filtered'] = visits['DC_dispo'].replace(rare_categories, 'Other')
```

```python
# Step 1: Count patients from "Start" to ED_dispo, and from ED_dispo to DC_dispo
start_to_ed = visits['ED_dispo'].value_counts().reset_index()
start_to_ed.columns = ['ED_dispo', 'count']
ed_to_dc = visits.groupby(['ED_dispo', 'DC_dispo_filtered']).size().reset_index(name='count')
```

```python
# Step 2: Create unique list of all nodes
all_nodes = ['Intake'] + list(start_to_ed['ED_dispo'].unique()) + list(ed_to_dc['DC_dispo_filtered'].unique())
node_indices = {label: i for i, label in enumerate(all_nodes)}
node_indices

# Step 3: Define links from Start → ED_dispo
start_links = {
    'source': [node_indices['Intake']] * len(start_to_ed),
    'target': [node_indices[ed] for ed in start_to_ed['ED_dispo']],
    'value': list(start_to_ed['count']),
}
```

```python
# Step 4: Define links from ED_dispo → DC_dispo
ed_links = {
    'source': [node_indices[ed] for ed in ed_to_dc['ED_dispo']],
    'target': [node_indices[dc] for dc in ed_to_dc['DC_dispo_filtered']],
    'value': list(ed_to_dc['count']),
}

# Step 5: Combine both sets of links
source = start_links['source'] + ed_links['source']
target = start_links['target'] + ed_links['target']
value = start_links['value'] + ed_links['value']
```

```python
# Step 6: Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=all_nodes
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
)])

fig.update_layout(title_text="Patient Flow: ED Disposition to Final Disposition", font_size=10)
fig.show()
```
